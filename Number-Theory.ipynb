# [Primality Test](https://en.wikipedia.org/wiki/Primality_test)

A primality test is an algorithm for determining whether an input number is prime. Among other fields of mathematics, it is used for cryptography. Unlike integer factorization, primality tests do not generally give prime factors, only stating whether the input number is prime or not. Factorization is thought to be a computationally difficult problem, whereas primality testing is comparatively easy (its running time is polynomial in the size of the input). Some primality tests prove that a number is prime, while others like Millerâ€“Rabin prove that a number is composite. Therefore, the latter might more accurately be called compositeness tests instead of primality tests.



```python
from math import isqrt
def is_prime(n: int) -> bool:
    if n <= 3:
        return n > 1
    if n % 2 == 0 or n % 3 == 0:
        return False
    limit = isqrt(n)
    for i in range(5, limit+1, 6):
        if n % i == 0 or n % (i+2) == 0:
            return False
    return True
```

```python
crivo = list(filter(is_prime,list(range(1000))))
print(crivo)
```

# [Goldbach Conjecture](https://en.wikipedia.org/wiki/Goldbach's_conjecture)
Goldbach's conjecture is one of the oldest and best-known unsolved problems in number theory and all of mathematics. It states that every even natural number greater than 2 is the sum of two prime numbers.

The conjecture has been shown to hold for all integers less than $4 \times 10^{18}$ but remains unproven despite considerable effort. 

```python
import matplotlib.pyplot as plt
def goldbach(crivo,n):
    assert(n % 2 == 0 and n > 2)
    for x in crivo:
        for y in crivo:
            if x + y == n:
                return x,y
    return "O goldbach estava errado"
```

# [Collatz Conjecture](https://en.wikipedia.org/wiki/Collatz_conjecture)
The Collatz conjecture is one of the most famous unsolved problems in mathematics. The conjecture asks whether repeating two simple arithmetic operations will eventually transform every positive integer into 1. It concerns sequences of integers in which each term is obtained from the previous term as follows: if the previous term is even, the next term is one half of the previous term. If the previous term is odd, the next term is 3 times the previous term plus 1. The conjecture is that these sequences always reach 1, no matter which positive integer is chosen to start the sequence. 

```python
def collatz(n):
    lista = []
    while n != 1:
        lista.append(n)
        if n % 2 == 1:
            n = 3 * n + 1
        else:
            n = n // 2
    return lista
```

```python
collatz(5)
```

```python
import matplotlib.pyplot as plt
for n in range(1,100):
    plt.plot(collatz(n))
plt.show()
```
