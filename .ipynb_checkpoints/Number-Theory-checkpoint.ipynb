```python
import math
def is_prime(x):
    lista = []
    for n in range(2,x):
        lista.append(n)
        if x % n == 0:
            return False
    print(lista)
    return True
def is_prime2(x):
    
    for n in range(2, math.ceil(math.sqrt(x))+1):
        lista.append(n)
        if x % n == 0:
            return False
    #print(lista)
    return True
```

```python
crivo = [2]+list(filter(is_prime2,list(range(1000))))[2:]
print(crivo)
```

```python
def goldbach(crivo,n):
    for x in crivo:
        for y in crivo:
            if x + y == n:
                return x,y
    return "O goldbach estava errado"
```

```python
goldbach(crivo,72)
```

```python
def collatz(n):
    lista = []
    while n != 1:
        lista.append(n)
        if n % 2 == 1:
            n = 3 * n + 1
        else:
            n = n // 2
    return lista
```

```python
collatz(5)
```

```python
import matplotlib.pyplot as plt
for n in range(1,100):
    plt.plot(collatz(n))
plt.show()
```
